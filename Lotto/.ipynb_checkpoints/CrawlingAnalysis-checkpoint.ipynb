```python
from bs4 import BeautifulSoup
import urllib.request
import os
import ssl
from urllib.request import urlopen
import pandas as pd

# 로또 메인 페이지
mainUrl = 'http://dhlottery.co.kr/common.do?method=main'
context = ssl._create_unverified_context()
result = urlopen(mainUrl, context=context)
soup = BeautifulSoup(result.read(), "lxml")
# soup = BeautifulSoup(urllib.request.urlopen(mainUrl).read(), "lxml")
    
# 마지막 회차 조회
def getLastSeries(soup=soup):
    return int(soup.find(id="lottoDrwNo").text)

def GetLottoHistoryToCsv(soup):   
    lastSeries = getLastSeries(soup)
    outputFileName = 'data/lotto_result.csv'

    lastNum = int(pd.read_csv(outputFileName)["1"].max())
             
    with open(outputFileName, 'a') as f:
        for i in range(lastNum+1, lastSeries+1):
            url = 'http://dhlottery.co.kr/gameResult.do?method=byWin&drwNo=' + str(i)
            soup = BeautifulSoup(urllib.request.urlopen(url).read(), "lxml")

            result = str(soup.find(id="desc")['content'])
            ret =  result.split('당첨번호')

            nubmerList = result.split('당첨번호')[1].split('.')[0].strip().split('+')[0].split((','))
            n1, n2, n3, n4, n5, n6 = nubmerList
            bonusNumber = result.split('당첨번호')[1].split('.')[0].strip().split('+')[1]
            totalCnt = result.split('당첨번호')[1].split('.')[1].split(('총'))[1].split(',')[0].strip().replace('명', '').replace(',', '')
            price = result.split('당첨번호')[1].split('.')[1].split(('총'))[1].split('당첨금액')[1].strip().replace('원', '').replace(',', '')

            print(str(i) + '회 당첨번호 :', nubmerList, '보너스 번호 :', bonusNumber, '1등 수 :', totalCnt, '인당 당첨금액 :', price)
            f.writelines(str(i) + ',' + n1 + ',' + n2 + ',' + n3 + ',' + n4 + ',' + n5 + ',' + n6 + ',' + bonusNumber + ',' + totalCnt + ',' + price + '\n')
    
    print("완료")
        
GetLottoHistoryToCsv(soup)
```

```python
import pandas as pd

def preProcess():
    c = ['회차', 'n1', 'n2', 'n3', 'n4', 'n5', 'n6', 'nb', '당첨자수', '인당 당첨금액']
    df_result = pd.read_csv('data/lotto_result.csv', header=None)
    df_result.columns = c
    df_result["sum"] = df_result["n1"]+df_result["n2"]+df_result["n3"]+df_result["n4"]+df_result["n5"]+df_result["n6"]
    
    return df_result
df_result = preProcess()
```

```python
from numpy.random import choice

def getNumber(thisSeries, n):
    vc = df_result[df_result["회차"] < thisSeries]["n"+str(n)].value_counts().sort_index()
#     num = choice(vc.index, 1, vc.tolist())[0]
    num = choice(vc.index, 1, [1 for _ in range(len(vc))])[0]
    return num

def generateNumbers(thisSeries, buyCnt):
    numSet = []
    for i in range(1, buyCnt+1):
        numSubSet = []
        while len(numSubSet) < 6:
            if getNumber(thisSeries, n) in numSubSet:
                pass
            else:
                numSubSet.append(getNumber(thisSeries, n))
            
        numSet.append(sorted(numSubSet))
    return numSet
```

```python
import os

def purchaseLotto(buyCnt=3):
    thisSeries = df_result["회차"].max()+1
    purchaseFileName = 'purchase/' + str(thisSeries) + '.csv'

    if os.path.exists(purchaseFileName):
        print("구매내역 존재!")
        return
#         os.remove(purchaseFileName)

    with open(purchaseFileName, 'a') as f:
        f.writelines(str(thisSeries) + ":" + str(generateNumbers(thisSeries, buyCnt)))

purchaseLotto(3)
```

```python

def simulation(fromSeries=2, buyCnt=3):
    pl = generateNumbers(fromSeries, 3)
    df_a = df_result[df_result["회차"] == fromSeries][["n1", "n2", "n3", "n4", "n5", "n6"]].T
    df_a.columns = ["answer"]
    answer = df_a["answer"].tolist()
    bonus_number = int(df_result[df_result["회차"] == fromSeries]["nb"])

    for l in pl:
        matchCnt = set(l).intersection(answer).__len__()
        matchNum = set(l).intersection(answer)
        if matchCnt < 3:
#             if matchCnt == 0:
#                 print('꽝(', matchCnt, '개 일치', ')', ' :', l)
#             else:
#                 print('꽝(', matchCnt, '개 일치', ')', ' :', l, matchNum)
            hitCnt[matchCnt] += 1
        elif matchCnt == 3:
#             print('5등 :', l, matchNum)
            hitCnt['5등'] += 1
        elif matchCnt == 4:
#             print('4등 :', l, matchNum)
            hitCnt['4등'] += 1
        elif matchCnt == 5:
            if bonus_number in l:
#                 print('2등 :', l, matchNum)
                hitCnt['2등'] += 1
            else:
#                 print('3등 :', l, matchNum)
                hitCnt['3등'] += 1
        elif matchCnt == 6:
            print('*********1등********* :', l, matchNum)
            hitCnt['1등'] += 1
    return hitCnt
```
