```python
df = pd.read_csv(dataDir+"005930"+".csv")
```

```python
df["diff_start"] = df["start"].diff()
df["diff_close"] = df["close"].diff()
df["diff_min"] = df["max"].diff()
df["diff_max"] = df["max"].diff()
df["diff_amount"] = df["amount"].diff()

df.tail()
```
