# 데이터 불러오기

```python
import json
import pandas as pd

with open("data/20210411.json", encoding="utf-8") as json_file:
    json_data = json.load(json_file)
    newsDf = pd.DataFrame(json_data)
newsDf
```

# 전처리

```python
import re
def CleanArticle(x):
    x = x.replace("// flash 오류를 우회하기 위한 함수 추가\nfunction _flash_removeCallback() {}\n\n", "").strip()
    return re.sub(r'[^ ㄱ-ㅣ가-힣A-Za-z]', '', x)

newsDf["body"] = newsDf["body"].apply(CleanArticle)
newsDf["title"] = newsDf["title"].apply(CleanArticle)
```

# 특정 기사에만 언급되는 핵심 단어 추출(TF-IDF)

## 형태소 분석

```python
from konlpy.tag import Hannanum

def GetWordTag(word):
    cls = Hannanum()
    return cls.pos(word)
```

```python
newsDict = dict()
print("시작")
for i in newsDf.index:
    newsDict[newsDf.iloc[i]["id"]] = GetWordTag(newsDf.iloc[i]["title"])
print("종료")
```

```python
newsDict
```

## 형태소별  Count

```python
import collections
list(map(collections.Counter, list(map(lambda x: list(map(lambda y:y[0], x)), newsDict.values()))))
```

## Word set 만들기

```python
wordSet = set([e[0] for w in newsDict.values() for e in w if len(e[0]) > 1 and e[1] in ('N')])
```

## 기사별 형태소가 wordSet에 있는지 체크

```python
collections.Counter([w[0] for w in newsDict.get('sid1=101&sid2=260&oid=001&aid=0012320653')])
```

```python
print(list(map(lambda k: collections.Counter([w[0] for w in newsDict.get('sid1=101&sid2=260&oid=001&aid=0012320653')])[k], wordSet)))
```

```python
result = []
for key, value in newsDict.items():
#     print(list(map(lambda k: collections.Counter([w[0] for w in newsDict.get(key)])[k], wordSet)))
#     df = df.append(pd.= Series(list(map(lambda k: collections.Counter([w[0] for w in newsDict.get(key)])[k], wordSet))), ignore_index=True)
    result.append([key] + list(map(lambda k: collections.Counter([w[0] for w in newsDict.get(key)])[k], wordSet)))
```

```python
countDf = pd.DataFrame(result)
countDf.columns = ["id"] + list(wordSet)
```
