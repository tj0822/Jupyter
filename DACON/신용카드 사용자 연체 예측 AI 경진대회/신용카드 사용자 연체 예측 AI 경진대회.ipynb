```python
def preprocess(df=None):
    global train
    df.drop('FLAG_MOBIL', axis=1, inplace=True)
    categoryFeatures = ['gender', 'car', 'reality', 'income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type']
    for column in df.columns:
        if column in categoryFeatures:
            if len(df[column].unique()) == 2:
                df[column] = list(map(lambda x: 0 if ((x == "F") or (x == "N") )else 1, list(df[column])))
            else:
                df = pd.concat([df, pd.get_dummies(df[column])], axis=1)
                df.drop(column, inplace=True, axis=1)
        else:
            df[column] = list(map(lambda x: x * -1 if (x < 0) else x, list(df[column])))
            if column not in ['index', 'credit'] and len(df[column].unique()) > 2:
                df[column] = df[column].apply(lambda x: (x-np.mean(trainDf[column]))/np.std(trainDf[column]), list(df[column]))
        
    c = list(df.columns).copy()
    if 'credit' in c:
        c.remove('credit')
        c.append('credit')
    df = df[c]       
    return df
train = preprocess(trainDf)
test = preprocess(testDf)
```

## Trainset 교차검증

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred, average=None)
    recall = recall_score(y_test, pred, average=None)
    
    print(confusion)
    print('accuracy :{0:.4f}, precision :{1:.4f}, recall :{2:.4f}'.format(accuracy, precision, recall))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, cross_validate
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
knn_clf = KNeighborsClassifier(n_neighbors=3)
lgbm_clf = LGBMClassifier(n_estimators=1000)
vo_clf = VotingClassifier(estimators=[('LGBMC', lgbm_clf), ('KNN', knn_clf), ('RF', rf_clf), ('DT', dt_clf)],
                          voting='soft')

# clf = vo_clf
# clf.fit(X_train, y_train)

clf = lgbm_clf
clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='logloss', eval_set=[(X_test, y_test)], verbose=True)

lr_pred = clf.predict(X_test)
lr_pred_proba = clf.predict_proba(X_test)[:, 1]
scores = cross_val_score(clf, X_train, y_train, scoring='accuracy', cv=5)
print('교차 검증별 정확도 : ', np.round(scores, 4))
print('평균 검증 정확도 : ', np.round(np.mean(scores), 4))
```
